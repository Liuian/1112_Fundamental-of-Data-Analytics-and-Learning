## one-hot encoding the categorical features

````python
columns = ['Pclass', 'Sex', 'Cabin', 'Embarked']
````

````python
#將categorical features的值做 one-hot encoding
def category_onehot_multcols(multcolumns):
    df_final=combined_df
    i=0  
    for fields in multcolumns:
        print(fields)
        #將 final_df 中的欄位 fields 的值做 one-hot encoding，並且移除掉每個欄位的第一個值，以避免多重共線性的問題。
        df1=pd.get_dummies(combined_df[fields],drop_first=True, dtype = bool) 
        combined_df.drop([fields],axis=1,inplace=True)
        if i==0:
            df_final=df1.copy()
        else:
            df_final=pd.concat([df_final,df1],axis=1)
        i=i+1
    df_final=pd.concat([combined_df,df_final],axis=1) 
    return df_final
````

````python
combined_df = category_onehot_multcols(columns)
````

````python
combined_df =combined_df.loc[:,~combined_df.columns.duplicated()]
````
